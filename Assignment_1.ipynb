## Q4)  Two Dice are rolled, find the probability that sum is

a)Equal to 1

b)Less than or equal to 4

c)Sum is divisible by 2 and  3

```python
Number_of_outcomes_for_each_Die = 6
Number_of_Die = 2

Total_number_of_possible_outcomes = Number_of_outcomes_for_each_Die ** Number_of_Die
print(Total_number_of_possible_outcomes)

a,b,c = 0,0,0

die1 = [1,2,3,4,5,6]
die2 = [1,2,3,4,5,6]

for i in die1:
    for j in die2:
        z= i+j
        if z == 1:
            a =a+1
        if z <= 4:
            b = b+1
        if z%2 == 0 and z%3 == 0:
            c = c+1
print(a,b,c)

Probability_of_a = a/Total_number_of_possible_outcomes
Probability_of_b = b/Total_number_of_possible_outcomes
Probability_of_c = c/Total_number_of_possible_outcomes

print("Probability of Equal to 1 is", Probability_of_a)
print("Probability_of_Less than or equal to 4 is", Probability_of_b)
print("Probability_of_Sum is divisible by 2 and 3 is",Probability_of_c )
```

## Q5)  A bag contains 2 red, 3 green and 2 blue balls. Two balls are drawn at random. What is the probability that none of the balls drawn is blue?

```python
R = 2
G = 3
B = 2

total_balls = R + G + B
total_number_of_outcomes = (total_balls * (total_balls - 1)) / 2*1

outcomes_in_both_balls_are_not_blue = ((R + G) * ((R + G) - 1)) / 2*1

probability_of_both_balls_are_not_blue = outcomes_in_both_balls_are_not_blue / total_number_of_outcomes

print("Probability_of_both_balls_are_not_blue is", probability_of_both_balls_are_not_blue)
```

##Q11)  Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?

```python
n = 2000                                                 # sample size
s = 30                                                   # Standard_Deviation_of_sample
x = 200                                                  # sample mean

# 94% Confidence Interval
t1 = stats.t.ppf(0.03,n-1)                                # t-score
# (s / √n) is already calculated i.e 211.29 and let it be 'm'
m = (s / math.sqrt(n) )
print(x+(t1*m), x-(t1*m))

# 98% Confidence Interval
t2 = stats.t.ppf(0.01,n-1)                                # t-score
# (s / √n) is already calculated i.e 211.29 and let it be 'm'
m = (s / math.sqrt(n) )
print(x+(t2*m), x-(t2*m))

# 96% Confidence Interval
t3 = stats.t.ppf(0.02,n-1)                                # t-score
# (s / √n) is already calculated i.e 211.29 and let it be 'm'
m = (s / math.sqrt(n) )
print(x+(t3*m), x-(t3*m))
```

###  Q 24)   A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days

```python
population_mean = 270
sample_mean = 260
sample_std_dev = 90
sample_size = 18

# Calculate the t-score
t_score = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

# Degrees of freedom (n - 1)
degrees_of_freedom = sample_size - 1

# Calculate the probability using the t-distribution
probability = stats.t.cdf(t_score, df=degrees_of_freedom)

print("Probability:", probability)
```
